```python
sns.countplot(x="gender",data=df,hue="Churn")
plt.title("Count of Customer by Churn on the basis of Gender in give data", fontsize=15,color="pin")
plt.show()
```

```python
# Create a countplot
sns.countplot(x="SeniorCitizen", data=df, hue="Churn")

# Get the current figure and axes
ax = plt.gca()

# Calculate the counts for each combination of SeniorCitizen and Churn
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Plot the stacked bars manually
counts.plot(kind='bar', stacked=True, ax=ax, color=sns.color_palette("Set2", n_colors=len(counts.columns)))

# Customize the plot
plt.title("Count of Customer by Churn on the basis of SeniorCitizen", fontsize=15, color="green")
plt.xlabel('SeniorCitizen')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep the x-ticks horizontal
plt.show()
```

```python
# Columns to visualize
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection']

# Number of columns to visualize
n_cols = len(columns)

# Set up the figure and axes
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  # Adjust rows/cols based on your data
axes = axes.flatten()  # Flatten the 2D array of axes for easier indexing

# Iterate through each column and create a countplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(f'Countplot of {col}', fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Count', fontsize=10)
    axes[i].tick_params(axis='x', rotation=30)  # Rotate x-ticks if needed

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
